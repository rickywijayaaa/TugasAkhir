##Fungsi Wrapper

Agar hasil output lebih readable karen tidak terlalu panjang ke samping

```python
import textwrap

def wrap_text(text, width=80):
    """
    Membungkus teks agar maksimal memiliki panjang 'width' karakter per baris,
    namun tetap mempertahankan format baris baru (\n) yang sudah ada.
    """
    if not text:
        return ""

    # Pisahkan berdasarkan baris baru yang sudah ada di knowledge base
    lines = text.split('\n')

    # Bungkus setiap baris agar tidak melebihi batas karakter (misal: 80)
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

    # Gabungkan kembali
    return '\n'.join(wrapped_lines)
```

## Rule-based Chatbot using regex

Referensi tambahan:
1. https://www.w3schools.com/python/python_regex.asp
2. https://docs.python.org/3/library/re.html

```python
import re

# ==========================================
# DEFINISI POLA REGEX (RULES)
# ==========================================

rules = [
    # Layanan Kemahasiswaan
    (re.compile(r".*\b(layanan|fasilitas)\b.*", re.I), JAWABAN_LAYANAN),

    # Asrama
    (re.compile(r".*\b(asrama)\b.*", re.I), JAWABAN_ASRAMA),

    # Bimbingan & Konseling
    (re.compile(r".*\b(bimbingan|konseling|bk|psikis|mental)\b.*", re.I), JAWABAN_TUJUAN_BK),

    # Organisasi Kemahasiswaan
    (re.compile(r".*\b(organisasi|ormawa|ukm|himpunan)\b.*", re.I), JAWABAN_ORGANISASI),

    # Mahasiswa Berprestasi
    (re.compile(r".*\b(berprestasi|mapres)\b.*", re.I), JAWABAN_MAPRES),

    # Pertanyaan spesifik tentang Beasiswa (Urutan berpengaruh)
    (re.compile(r".*\b(berhenti|dihentikan|dicabut|putus|kondisi|penghentian)\b.*\b(beasiswa)\b.*", re.I), JAWABAN_HENTI_BEASISWA),
    (re.compile(r".*\b(kewajiban|tugas|wajib|harus)\b.*\b(beasiswa)\b.*", re.I), JAWABAN_KEWAJIBAN_BEASISWA),
    (re.compile(r".*\b(syarat|kriteria)\b.*\b(beasiswa)\b.*", re.I), JAWABAN_SYARAT_BEASISWA),
    (re.compile(r".*\b(beasiswa)\b.*", re.I), "Terkait beasiswa, apakah kamu ingin tahu tentang syarat penerima, kewajiban, atau kondisi penghentiannya?"),

    # Pertanyaan spesifik tentang Ganesha
    (re.compile(r".*\b(bidang)\b.*\b(ganesha)\b.*", re.I), JAWABAN_BIDANG_GANESHA),
    (re.compile(r".*\b(kategori|jenis|macam)\b.*\b(ganesha)\b.*", re.I), JAWABAN_KATEGORI_GANESHA),
    (re.compile(r".*\b(ganesha)\b.*", re.I), "Untuk Penghargaan Ganesha, apakah kamu ingin tahu tentang kategori penghargaannya atau bidang yang diakui?"),
]

# Fallback
fallback = "Maaf, aku kurang paham pertanyaanmu. Coba tanyakan seputar layanan kemahasiswaan, aturan asrama, bimbingan konseling, organisasi kemahasiswaan, mahasiswa berprestasi, beasiswa, atau penghargaan Ganesha di ITB."

# Variabel untuk menyimpan ingatan/konteks percakapan
current_context = ""

def find_answer_with_context(text: str) -> str:
    global current_context

    # 1. Cek apakah ada konteks percakapan sebelumnya (Memori)
    if current_context == "beasiswa":
        if re.match(r".*\b(syarat|kriteria)\b.*", text, re.I):
            current_context = "" # Reset memori
            return JAWABAN_SYARAT_BEASISWA
        elif re.match(r".*\b(kewajiban|tugas|wajib|harus)\b.*", text, re.I):
            current_context = "" # Reset memori
            return JAWABAN_KEWAJIBAN_BEASISWA
        elif re.match(r".*\b(berhenti|dihentikan|dicabut|putus|penghentian|kondisi)\b.*", text, re.I):
            current_context = "" # Reset memori
            return JAWABAN_HENTI_BEASISWA

    elif current_context == "ganesha":
        if re.match(r".*\b(bidang)\b.*", text, re.I):
            current_context = ""
            return JAWABAN_BIDANG_GANESHA
        elif re.match(r".*\b(kategori|jenis|macam)\b.*", text, re.I):
            current_context = ""
            return JAWABAN_KATEGORI_GANESHA

    # 2. Jika tidak ada konteks, cari pakai rules biasa
    for pat, ans in rules:
        if pat.match(text):
            # Jika jawaban memancing pertanyaan lanjutan, simpan ke memori
            if "Terkait beasiswa" in ans:
                current_context = "beasiswa"
            elif "Untuk Penghargaan Ganesha" in ans:
                current_context = "ganesha"
            else:
                current_context = "" # Reset jika topiknya langsung dijawab

            return ans

    # Jika benar-benar tidak paham, kembalikan fallback
    return fallback

if __name__ == "__main__":
    print("Chatbot Peraturan Kemahasiswaan ITB. Ketik 'exit' untuk keluar.")
    while True:
        user = input("\nYou: ").strip()
        if not user or user.lower() in ("exit", "quit", "keluar"):
            print("Bot: Sampai jumpa!")
            break

        # Dapatkan jawaban asli
        jawaban = find_answer_with_context(user)

        # Cetak dengan format yang sudah dibungkus (maks 80 karakter per baris)
        print("Bot:\n" + wrap_text(jawaban, width=80))
```

## Rule-based Chatbot using NLTK
Sumber: https://www.analyticsvidhya.com/blog/2021/07/build-a-simple-chatbot-using-python-and-nltk/

```python
import nltk
from nltk.chat.util import Chat, reflections

# ==========================================
# DEFINISI POLA NLTK PAIRS
# Format: [ r"pola_regex_string", ["List_Jawaban"] ]
# ==========================================
pairs = [
    [
        r"(.*)(layanan|fasilitas)(.*)",
        [JAWABAN_LAYANAN]
    ],
    [
        r"(.*)(asrama)(.*)",
        [JAWABAN_ASRAMA]
    ],
    [
        r"(.*)(bimbingan|konseling|bk|psikis|mental)(.*)",
        [JAWABAN_TUJUAN_BK]
    ],
    [
        r"(.*)(organisasi|ormawa|ukm|himpunan)(.*)",
        [JAWABAN_ORGANISASI]
    ],
    [
        r"(.*)(berprestasi|mapres)(.*)",
        [JAWABAN_MAPRES]
    ],
    # Hierarki pertanyaan Beasiswa
    [
        r"(.*)(berhenti|dihentikan|dicabut|putus)(.*)beasiswa(.*)",
        [JAWABAN_HENTI_BEASISWA]
    ],
    [
        r"(.*)(kewajiban|tugas|wajib|harus)(.*)beasiswa(.*)",
        [JAWABAN_KEWAJIBAN_BEASISWA]
    ],
    [
        r"(.*)(syarat|kriteria)(.*)beasiswa(.*)",
        [JAWABAN_SYARAT_BEASISWA]
    ],
    [
        r"(.*)beasiswa(.*)",
        ["Terkait beasiswa, apakah kamu ingin tahu tentang 'syarat beasiswa', 'kewajiban beasiswa', atau 'kondisi penghentian beasiswa'?"]
    ],
    # Hierarki pertanyaan Ganesha
    [
        r"(.*)(bidang)(.*)ganesha(.*)",
        [JAWABAN_BIDANG_GANESHA]
    ],
    [
        r"(.*)(kategori|jenis|macam)(.*)ganesha(.*)",
        [JAWABAN_KATEGORI_GANESHA]
    ],
    [
        r"(.*)ganesha(.*)",
        ["Untuk Penghargaan Ganesha, silakan perjelas apakah kamu mencari 'kategori ganesha' atau 'bidang ganesha'?"]
    ],
    # Fallback
    [
        r"(.*)",
        ["Maaf, aku kurang paham pertanyaanmu. Coba tanyakan seputar layanan kemahasiswaan, aturan asrama, bimbingan konseling, organisasi kemahasiswaan, mahasiswa berprestasi, beasiswa, atau penghargaan Ganesha di ITB."]
    ]
]

# Inisialisasi NLTK Chat
chatbot = Chat(pairs, reflections)

# Fungsi interaksi
def start_nltk_chat():
    print("NLTK Chatbot Peraturan Kemahasiswaan ITB. Ketik 'exit' untuk keluar.")
    while True:
        user_input = input("\nYou: ").strip()
        if not user_input or user_input.lower() in ("exit", "quit", "keluar"):
            print("Bot: Sampai jumpa!")
            break

        # Dapatkan respons otomatis menggunakan objek NLTK Chat
        response = chatbot.respond(user_input)

        # Cetak dengan format yang sudah dibungkus (maks 80 karakter per baris)
        print("Bot:\n" + wrap_text(response, width=80))

if __name__ == "__main__":
    start_nltk_chat()
```

## Chatbot WITH LLM (tanpa RAG)

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

MODEL_ID = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

# Load
tok = AutoTokenizer.from_pretrained(MODEL_ID)
pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(
        MODEL_ID,
        torch_dtype="auto",
        device_map="auto"
    ),
    tokenizer=tok
)

def chat_once(user_msg: str) -> str:
    messages = [
        {"role": "system", "content":
                "Anda asisten akademik Institut Teknologi Bandung yang menjawab berdasarkan konteks yang diberikan. "
                "Jawab SINGKAT, JELAS, dan HANYA dari konteks. "
                "Jika tidak ada informasi relevan, balas: 'Tidak ditemukan dalam dokumen.' "
                "Jangan sertakan rujukan/sumber dalam jawaban."},
        {"role": "user", "content": user_msg},
    ]
    prompt = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    out = pipe(
        prompt,
        max_new_tokens=256,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        eos_token_id=tok.eos_token_id,
    )[0]["generated_text"]

    reply = out[len(prompt):].strip()
    return reply

if __name__ == "__main__":
    print("HF Chatbot (TinyLlama). ketik 'exit' untuk keluar.")
    while True:
        u = input("You: ")
        if u.strip().lower() == "exit":
            break
        print("Bot:", chat_once(u))
```

```python
# Cell 2 - Setup PDF reader (Colab) — PDF-only tanpa OCR
import os, re
from pypdf import PdfReader

# Ganti ke path PDF kamu di Colab
PDF_PATH = "/content/Peraturan_Kemahasiswaan_2022.pdf"

def verify_pdf(path: str) -> bool:
    if not os.path.isfile(path):
        raise FileNotFoundError(f"Tidak menemukan PDF di: {path}")
    try:
        reader = PdfReader(path)
        n_pages = len(reader.pages)
        if n_pages == 0:
            raise ValueError("PDF tidak memiliki halaman.")
        print(f"PDF OK — jumlah halaman: {n_pages}")
        return True
    except Exception as e:
        raise RuntimeError(f"Gagal membaca PDF: {e}")

def extract_pages(path: str):
    """Ekstrak teks per halaman (tanpa OCR). Halaman kosong di-skip."""
    reader = PdfReader(path)
    pages = []
    for i, page in enumerate(reader.pages, start=1):
        txt = page.extract_text() or ""
        # Normalisasi ringan
        txt = txt.replace("\u00ad", "")     # soft hyphen
        txt = txt.replace("-\n", "")        # sambung kata yang terpotong
        txt = re.sub(r"\s+", " ", txt).strip()
        if txt:
            pages.append({"page": i, "text": txt})
    print(f"Halaman dengan teks: {len(pages)}")
    return pages

# Verifikasi & contoh ekstraksi
verify_pdf(PDF_PATH)
pages_text = extract_pages(PDF_PATH)
if pages_text:
    print("Cuplikan halaman 1 (200 karakter):")
    print(pages_text[0]["text"][:200])
```

```python
# Cell 4 - PDF processing (tanpa OCR)

def clean_text(s: str) -> str:
    """Bersihkan teks dari karakter aneh & normalisasi spasi/baris."""
    s = (s or "")
    s = s.replace("\x00", " ").replace("\ufb01", "fi").replace("\ufb02", "fl")  # ligatures umum
    s = s.replace("\u00ad", "")        # soft hyphen
    s = s.replace("-\n", "")           # sambung kata terpotong di line break
    s = re.sub(r"\n{3,}", "\n\n", s)   # kompres newline berlebih
    s = re.sub(r"[ \t]+", " ", s)      # kompres spasi/tab
    return s.strip()

def read_pdf_text(path: str, min_chars: int = 50) -> Tuple[str, List[Tuple[int, str]]]:
    print(f"Membaca PDF: {os.path.basename(path)}")

    reader = PdfReader(path)
    pages: List[Tuple[int, str]] = []

    for i, p in enumerate(reader.pages, start=1):
        raw = p.extract_text() or ""
        t = clean_text(raw)
        pages.append((i, t))
        print(f"  Halaman {i}: {len(t)} karakter")

    # Peringatan untuk halaman yang tampak kosong/scan
    suspect = [i for (i, t) in pages if len(t) < min_chars]
    if suspect:
        print(f"Peringatan: {len(suspect)} halaman terdeteksi minim teks {suspect}. "
              f"Kemungkinan berbasis gambar. Mode ini PDF-only, OCR dinonaktifkan.")

    full = "\n\n".join([f"[page {i}]\n{t}" for i, t in pages])
    total_chars = sum(len(t) for _, t in pages)
    print(f"Total: {len(pages)} halaman, {total_chars} karakter (tanpa OCR)")

    return full, pages

# Contoh pakai (opsional):
full_text, pages_list = read_pdf_text(PDF_PATH)
print(full_text[:500])
```

```python
# Cell 5 - Path file PDF (PDF-only)
import os

PDF_PATH = "/content/Peraturan_Kemahasiswaan_2022.pdf"

def set_pdf_list(path: str):
    if os.path.isfile(path):
        print(f"File PDF ditemukan: {os.path.basename(path)}")
        return [path]
    else:
        print(f"File tidak ditemukan: {path}")
        print("Pastikan path file sudah benar.")
        return []

PDFS = set_pdf_list(PDF_PATH)
```

```python
# Cell 6 - Load & proses dokumen PDF
import re, os

all_docs = []

# BAB
RE_BAB = re.compile(
    r'(?m)^\s*(BAB\s+[IVXLC]+\s*\n\s*[^\n]+)\s*\n(.*?)(?=^\s*BAB\s+[IVXLC]+|\Z)',
    flags=re.DOTALL
)

# BAGIAN
RE_BAGIAN = re.compile(
    r'(?m)^\s*(Bagian\s+[A-Za-z]+\s*\n\s*[^\n]+)\s*\n(.*?)(?=^\s*Bagian\s+[A-Za-z]+|\Z)',
    flags=re.DOTALL
)

# PASAL
RE_PASAL = re.compile(
    r'(?m)^\s*(Pasal\s+\d+)\s*\n(.*?)(?=^\s*Pasal\s+\d+|\Z)',
    flags=re.DOTALL
)

# AYAT
RE_AYAT = re.compile(
    r'(?m)^\s*(\(\d+\))\s*(.+?)(?=\n\s*\(\d+\)|\Z)',
    flags=re.DOTALL
)

for path in PDFS:

    print(f"\n=== Proses: {os.path.basename(path)} ===")

    full_text, page_texts = read_pdf_text(path)
    source_name = os.path.basename(path)
    full_text = re.sub(r'\r', '', full_text)
    full_text = re.sub(r'[ \t]+', ' ', full_text)

    for bab_match in RE_BAB.finditer(full_text):

        bab_title = bab_match.group(1).strip()
        bab_body = bab_match.group(2).strip()

        all_docs.append({
            "text": f"{bab_title}\n{bab_body}",
            "level": "BAB",
            "bab": bab_title,
            "source": source_name
        })

        for bagian_match in RE_BAGIAN.finditer(bab_body):

            bagian_title = bagian_match.group(1).strip()
            bagian_body = bagian_match.group(2).strip()

            all_docs.append({
                "text": f"{bab_title}\n{bagian_title}\n{bagian_body}",
                "level": "BAGIAN",
                "bab": bab_title,
                "bagian": bagian_title,
                "source": source_name })

    for pasal_match in RE_PASAL.finditer(full_text):

        pasal_title = pasal_match.group(1).strip()
        pasal_body = pasal_match.group(2).strip()

        all_docs.append({
            "text": f"{pasal_title}\n{pasal_body}",
            "level": "PASAL",
            "pasal": pasal_title,
            "source": source_name
        })

        for ayat_match in RE_AYAT.finditer(pasal_body):

            ayat_num = ayat_match.group(1)
            ayat_text = ayat_match.group(2).strip()

            all_docs.append({
                "text": f"{pasal_title} {ayat_num}\n{ayat_text}",
                "level": "AYAT",
                "pasal": pasal_title,
                "ayat": ayat_num,
                "source": source_name
            })

print("chunks:", len(all_docs))
```

```python
# Cell 7 - Setup embedding & FAISS index (PDF-only)
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

if not all_docs:
    raise ValueError("Tidak ada dokumen untuk di-embed. Pastikan Cell 6 menghasilkan 'all_docs' > 0.")

EMB_MODEL = "intfloat/multilingual-e5-base"
print(f"Memuat embedding model: {EMB_MODEL}")
embedder = SentenceTransformer(EMB_MODEL)

corpus = [d["text"] for d in all_docs]
print(f"Jumlah chunk untuk di-embed: {len(corpus)}")

# Encode ke vektor (cosine via normalize_embeddings=True → pakai IndexFlatIP)
X = embedder.encode(
    corpus,
    batch_size=16,
    normalize_embeddings=True,
    show_progress_bar=True
)
X = np.array(X, dtype="float32")

print("Membangun FAISS index (Inner Product / Cosine-normalized)...")
index = faiss.IndexFlatIP(X.shape[1])
index.add(X)

print(f"Index siap. Dimensi: {X.shape[1]}, Jumlah vektor: {index.ntotal}")
```

```python
# Cell 8 - Setup reranker (opsional tapi disarankan)
from sentence_transformers import CrossEncoder
import numpy as np

RERANKER = "BAAI/bge-reranker-base"
print(f"Memuat reranker: {RERANKER}")
reranker = CrossEncoder(RERANKER)  # otomatis pakai GPU jika tersedia (via torch)

def rerank_pairs(query: str, docs: List[Dict], topk: int = 5) -> List[Dict]:
    """
    Re-ranking doc hasil retrieval awal (mis. dari FAISS) dengan CrossEncoder.
    Mengembalikan hingga topk dokumen terbaik, menambahkan 'rerank_score'.
    """
    if not docs:
        return []
    k = min(topk, len(docs))
    pairs = [[query, d["text"]] for d in docs]
    scores = np.asarray(reranker.predict(pairs), dtype="float32")
    order = np.argsort(-scores)[:k]
    return [docs[i] | {"rerank_score": float(scores[i])} for i in order]
```

```python
# Cell 9 - Setup LLM
LLM = "Qwen/Qwen2.5-1.5B-Instruct"

device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(LLM, use_fast=True)
model = AutoModelForCausalLM.from_pretrained(
    LLM,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    low_cpu_mem_usage=True,
)
model.to(device)
model.eval()

# def make_messages(query: str, ctxs: List[Dict]) -> List[Dict]:
#     bullets = []
#     for c in ctxs:
#         snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:900]
#         bullets.append(f"- {snippet}")
#     context_block = "\n".join(bullets) if bullets else "(konteks kosong)"
#     return [
#         {
#             "role": "system",
#             "content": (
#                 "Anda asisten akademik yang menjawab berdasarkan konteks yang diberikan. "
#                 "Jawab SINGKAT, JELAS, dan HANYA dari konteks. "
#                 "Jika tidak ada informasi relevan, balas: 'Tidak ditemukan dalam dokumen.' "
#                 "Jangan sertakan rujukan/sumber dalam jawaban."
#             ),
#         },
#         {
#             "role": "user",
#             "content": (
#                 f"KONTEKS:\n{context_block}\n\n"
#                 f"PERTANYAAN:\n{query}\n\n"
#                 f"Jawab singkat dan jelas tanpa menyertakan rujukan."
#             ),
#         },
#     ]

@torch.inference_mode()
def generate_answer(messages, max_new_tokens=256):
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    input_len = inputs["input_ids"].shape[1]
    out = model.generate(
        **inputs,
        do_sample=False,
        max_new_tokens=max_new_tokens,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )[0]
    gen = out[input_len:]
    txt = tokenizer.decode(gen, skip_special_tokens=True).strip()
    return re.sub(r"^assistant\s*:\s*", "", txt, flags=re.I) or "Tidak ditemukan dalam dokumen."
```

```python
# Cell 10 - Retrieval & Chat khusus cerita "Kota Senja"

from dataclasses import dataclass
from typing import List, Dict, Optional
import numpy as np
import re, os

@dataclass
class Retrieved:
    text: str
    score: float
    bab: str
    pasal: str
    ayat: str
    source: str

PAT_PASAL = re.compile(r"Pasal\s+(\d+)", re.I)
PAT_AYAT = re.compile(r"\((\d+)\)")
PAT_BAB = re.compile(r"BAB\s+[IVXLC]+", re.I)

def boost_ids_from_query(query: str):

    boost_ids = []

    pasal_match = PAT_PASAL.search(query)
    ayat_match = PAT_AYAT.search(query)

    for i, d in enumerate(all_docs):

        text = d["text"]

        if pasal_match:
            if f"Pasal {pasal_match.group(1)}" in text:
                boost_ids.append(i)

        if ayat_match:
            if f"({ayat_match.group(1)})" in text:
                boost_ids.append(i)

    return boost_ids

def make_messages_academic(query: str, ctxs: List[Dict]) -> List[Dict]:
    bullets = []
    for c in ctxs:
        snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:900]
        bullets.append(f"- {snippet}")
    context_block = "\n".join(bullets) if bullets else "(konteks kosong)"
    return [
        {
            "role": "system",
            "content": (
                "Anda adalah asisten akademik Institut Teknologi Bandung yang dapat menjawab pertanyaan mengenai Peraturan Akademik"
                "Jawab SINGKAT, JELAS, dan HANYA berdasarkan konteks. "
                "Jika tidak ada informasi relevan di konteks, balas: 'Tidak ditemukan dalam Peraturan Akademik.' "
                "Jangan menambahkan rujukan/sumber dalam jawaban."
            ),
        },
        {
            "role": "user",
            "content": (
                f"KONTEKS:\n{context_block}\n\n"
                f"PERTANYAAN:\n{query}\n\n"
                f"Jawab singkat dan jelas tanpa menyertakan rujukan."
            ),
        },
    ]

def retrieve(query: str, k: int = 12, topk_final: int = 6) -> List[Retrieved]:
    # 1) Booster dari kode/nama pada query
    bonus_ids = boost_ids_from_query(query)

    # 2) Semantic top-k via FAISS
    qv = np.array(embedder.encode([query], normalize_embeddings=True), dtype="float32")
    D, I = index.search(qv, min(k, index.ntotal))
    I = I[0].tolist(); D = D[0].tolist()

    # 3) Gabungkan kandidat + skor awal
    cand, seen = [], set()
    for bid in bonus_ids + I:
        if bid in seen:
            continue
        seen.add(bid)
        d = all_docs[bid]
        base = D[I.index(bid)] if bid in I else 0.0
        sc = base
        if bid in bonus_ids:
          sc += 0.15
        cand.append(d | {"score": float(sc)})

    if not cand:
        return []

    # 4) Re-rank (CrossEncoder) → ambil topk_final
    reranked = rerank_pairs(query, cand, topk=topk_final)
    out: List[Retrieved] = []
    for c in reranked:
        out.append(Retrieved(
            text=c["text"],
            score=float(c.get("rerank_score", c["score"])),
            bab=c.get("bab"),
            pasal=c.get("pasal"),
            ayat=c.get("ayat"),
            source=c.get("source")
        ))
    return out

def chat(query: str) -> Dict:
    ctxs = retrieve(query)
    msgs = make_messages_academic(query, [c.__dict__ for c in ctxs])
    ans = generate_answer(msgs)
    cites = [
        {
            "bab": c.bab,
            "pasal": c.pasal,
            "ayat": c.ayat,
            "source": c.source
        }
        for c in ctxs
    ]
    if ans.strip().lower() == (query or "").strip().lower():
        ans = "Tidak ditemukan dalam Peraturan Akademik."
    return {"answer": ans, "contexts": cites}

print("="*60)
print("RAG Peraturan Akademik ITB — siap dipakai!")
print("="*60)
```

```python
# Cell 11 - Chat interaktif

def start_chat(show_sources: bool = False):
    """Mulai chat interaktif untuk naskah 'Sepuluh Sahabat ke Kota Senja'."""
    print("CHATBOT 'Peraturan Akademik ITB' AKTIF!")
    print("Tanya apa saja tentang naskah 'Peraturan Akademik ITB'.")
    print("Ketik 'exit' untuk keluar.")
    print("="*50)

    while True:
        try:
            q = input("\nAnda: ").strip()
        except EOFError:
            print("\nSesi dihentikan.")
            break

        if not q or q.lower() in {"exit", "quit", "q"}:
            print("Terima kasih!")
            break

        out = chat(q)
        print(f"\nBot: {out['answer']}")

        if show_sources and out.get("contexts"):
            print("\n[Sumber konteks]")
            for i, c in enumerate(out["contexts"], start=1):
                sec = c.get("section") or "—"
                pg  = c.get("page_hint")
                src = c.get("source") or ""
                pg_str = f"hal {pg}" if pg else "hal ?"
                print(f"{i}. {sec} ({pg_str}, {src})")

        print("-" * 50)

if __name__ == "__main__":
    # Ubah ke True jika ingin melihat sumber konteks yang dipakai
    start_chat(show_sources=False)
```
